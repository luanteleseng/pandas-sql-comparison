SELECT  Country, Continent, 2022 Population, 1970 Population, 2022 Population - 1970 Population as Growth  
FROM df  
WHERE Continent = "South America"

```python
query = df[['Country', 'Continent', '2022 Population', '1970 Population']] \
.assign(Growth = df['2022 Population'] - df['1970 Population'])

querySA = query[query['Continent'] == 'South America']
querySA
```

SELECT *  
FROM query   
WHERE Country IN ('Argentina', 'Brazil', 'Paraguay', 'Uruguay', 'Venezuela');

```python
querySAMerc = querySA.loc[querySA['Country'] \
            .isin(['Argentina', 'Brazil', 'Paraguay', 'Uruguay', 'Venezuela'])]
```

```python
querySAMerc = querySAMerc.assign(GrowthPercent = (((df['2022 Population'] - df['1970 Population'])/(df['1970 Population'])) * 100).round(2))
querySAMerc
```
